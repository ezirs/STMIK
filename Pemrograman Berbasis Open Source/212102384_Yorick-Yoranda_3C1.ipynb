```python
angka = int(input("Masukkan angka = "))
listGolongan = ''

def is_prima (x):
  for i in range(2, x):
    if x % i == 0:
      return False

  return True

def bilanganBulat(nilai):
    if (nilai % 2 == 0):
        return "Bilangan Bulat "
    else:
        return ""

def bilanganCacah(nilai):
    if (nilai >= 0):
        return "Bilangan Cacah "
    else:
        return ""

def bilanganNegatif(nilai):
    if (nilai < 0):
        return "Bilangan Negatif "
    else:
        return ""
    
def bilanganNol(nilai):
    if (nilai == 0):
        return "Bilangan Nol "
    else:
        return ""
    
def bilanganAsli(nilai):
    if (nilai > 0):
        return "Bilangan Asli "
    else:
        return ""

def bilanganGanjil(nilai):
    if (nilai % 2 == 0):
        return ""
    else:
        return "Bilangan Ganjil "
    
def bilanganGenap(nilai):
    if (nilai % 2 == 0):
        return "Bilangan Genap "
    else:
        return ""
    
def bilanganPrima(nilai):
    if (is_prima(nilai)):
        return "Bilangan Prima "
    else:
        return ""
    
def bilanganKomposit(nilai):
    if (is_prima(nilai) and nilai < 1):
        return ""
    else:
        return "Bilangan Komposit "


listGolongan = bilanganBulat(angka) + bilanganCacah(angka) + bilanganNegatif(angka) + bilanganNol(angka) + bilanganAsli(angka) + bilanganGanjil(angka) + bilanganGenap(angka) + bilanganPrima(angka) + bilanganKomposit(angka)
    
print("Angka %s tergolong %s" % (angka, listGolongan))
        
        
    
```

```python
kode = ['SP', 'PB', 'PA', 'DD', 'FF', 'NS']
namaMenu = ['Sayap', 'Paha Bawah', 'Paha Atas', 'Dada', 'French Fries', 'Nasi']
harga = [7000, 8000, 11000, 11000, 7000, 4000]


banyakJenis = int(input("Banyak Jenis : "))
totalHarga = 0

listHarga = list()
listBeli = list()
listBanyak = list()

i = 1
while i <= banyakJenis:
    print("Jenis ke - %s" % i)
    kodePotong = str(input("Kode Potong [SP/PB/PA/DD/FF/NS] : "))
    banyakPotong = int(input("Banyak Potong : "))
    idx = kode.index(kodePotong)
    getHarga = harga[idx]
    listHarga.append(getHarga)
    listBeli.append(namaMenu[idx])
    listBanyak.append(banyakPotong)
    totalHarga = totalHarga + (getHarga * banyakPotong)
    i += 1


print("FRIED CHICKEN LOCK DOWN")
print("-" * 60)
print('No.' + (' ' * 10) + 'Jenis' + (' ' * 8) + 'Harga' + (' ' * 7) + 'Banyak' + (' ' * 7) + 'Jumlah')
print('Potongan' + (' ' * 5) + 'Satuan' + (' ' * 7) + 'Beli' + (' ' * 21) + 'Harga')
print("-" * 60)

jumlahBayar = 0


i = 0
while i <= (len(listBeli) - 1):
    jumlahBayar = jumlahBayar + (listHarga[i] * listBanyak[i])
    print("%s" % i + (' ' * 12) + '%s' % listBeli[i] + (' ' * 11) + '%s' % listHarga[i] + (' ' * 6) + '%s' % listBanyak[i] + (' ' * 6) + 'Rp', (listHarga[i] * listBanyak[i]))
    i += 1

pajak = (10 * jumlahBayar) / 100
totalBayar = jumlahBayar + pajak
print("-" * 60)
print("Jumlah Bayar Rp. ", jumlahBayar)
print("Pajak 10%    Rp. ", pajak)
print("Total Bayar  Rp. ", totalBayar)
```
