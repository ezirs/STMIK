```python
class Mobil(object):
    
    def __init__(self, nomor_plat, warna, manufaktur, jenis, kecepatan):
        self.nomor_plat = nomor_plat
        self.warna = warna
        self.manufaktur = manufaktur
        self.jenis = jenis
        self.kecepatan = kecepatan
    
    def akselerasi(self):
        print("Akselerasi Mobil %s adalah %s" % (self.manufaktur, self.kecepatan))

    def belok_kiri(self):
        print("Mobil dengan plat %s Belok Kiri" % (self.nomor_plat))

    def belok_kanan(self):
        print("Mobil jenis %s dengan warna %s Belok Kanan" % (self.jenis, self.warna))

    def rem(self):
        print("Mobil %s rem dalam kecepatan %s" % (self.manufaktur, self.kecepatan))
    
mobil1_ku = Mobil("ABC111", "Hitam", "Mitsubishi", "Pajero Sport", "180km/h")
mobil1_ku.akselerasi()
mobil1_ku.belok_kiri()
mobil1_ku.belok_kanan()
mobil1_ku.rem()
print("\n")
mobil1_ku = Mobil("XYZ123", "Merah", "Toyota", "Innova", "120km/h")
mobil1_ku.akselerasi()
mobil1_ku.belok_kiri()
mobil1_ku.belok_kanan()
mobil1_ku.rem()
print("\n")
mobil1_ku = Mobil("DEF987", "Putih", "Daihatsu", "Sirion", "100km/h")
mobil1_ku.akselerasi()
mobil1_ku.belok_kiri()
mobil1_ku.belok_kanan()
mobil1_ku.rem()

```

```python
class Smartphone(object):
    
    def __init__(self, nama, layar, chipset, ram, rom, kamera_utama, kamera_depan, baterai, harga, volume):
        self.nama = nama
        self.layar = layar
        self.chipset = chipset
        self.ram = ram
        self.rom = rom
        self.kamera_utama = kamera_utama
        self.kamera_depan = kamera_depan
        self.baterai = baterai
        self.harga = harga
        self.volume = volume
    
    def powerOn(self):
        print("Anda menghidupkan hp %s" % (self.nama))

    def powerOff(self):
        print("Anda mematikan hp %s" % (self.nama))

    def volumeUp(self, volume_up):
        self.volume = self.volume + volume_up
        print("Anda menaikkan volume hp %s ke %s" % (self.nama, self.volume))

    def volumeDown(self, volume_down):
        self.volume = self.volume - volume_down
        print("Anda menurunkan volume hp %s ke %s" % (self.nama, self.volume))
        
    def getVolume(self):
        print("Volume saat ini : %s" % (self.volume))

    def detail(self):
        print("Nama : %s" % self.nama)
        print("Layar : %s" % self.layar)
        print("Chipset : %s" % self.chipset)
        print("RAM : %s" % self.ram)
        print("ROM : %s" % self.rom)
        print("Kamera Utama : %s" % self.kamera_utama)
        print("Kamera Depan : %s" % self.kamera_depan)
        print("Baterai : %s" % self.baterai)
        print("Harga : %s" % self.harga)
        print("Volume : %s" % self.volume)

def aksi(sp):
    sp.detail()
    i = 0
    while i < 1:
        aksi = str(input("Pilih [poweron, poweroff, volumeup, volumedown, getvolume, keluar] : "))
        if aksi == "poweron":
            sp.powerOn()
        elif aksi == "poweroff":
            sp.powerOff()
        elif aksi == "volumeup":
            vl = int(input("Masukkan volume : "))
            sp.volumeUp(vl)
        elif aksi == "volumedown":
            vl = int(input("Masukkan volume : "))
            sp.volumeDown(vl)
        elif aksi == "getvolume":
            sp.getVolume()
        elif aksi == "keluar":
            i += 1
        

hp = str(input("Select Object [sp1, sp2, sp3] : "))
if hp == "sp1":
    sp = Smartphone("Vivo V9 6 GB", "6.3 Inci (1080 x 2280 pixel)", "Qualcomm Snapdragon 660", "6 GB", "64 GB", "13 MP + 2 MP", "12 MP", "3.260 mAh", "Rp 4.200.000", 0)
    aksi(sp)
elif hp == "sp2":
    sp = Smartphone("Samsung Galaxy A8 Star", "6.3 Inci (2220 x 1080 pixel)", "Qualcomm Snapdragon 660", "4 GB", "64 GB", "24 MP + 16 MP", "24 MP", "3.700 mAh", "Rp 7.900.000", 0)
    aksi(sp)
elif hp == "sp3":
    sp = Smartphone("Xiaomi A2", "5.99 Inci (1080 x 2160 pixel)", "Qualcomm Snapdragon 660", "6 GB", "128 GB", "12 MP + 20 MP", "20 MP", "3.000 mAh", "Rp 4.700.000", 0)
    aksi(sp)
    
```
