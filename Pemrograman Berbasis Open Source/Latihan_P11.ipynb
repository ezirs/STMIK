```python
class distributor:
    def __init__(self, nama, nomorID, namaUnit, hargaBarang, jumlahTerjual, totalPendapatan):
        self.__setDistributor(nama, nomorID, namaUnit, hargaBarang, jumlahTerjual, totalPendapatan)
        
    def __setDistributor(self, nama, nomorID, namaUnit, hargaBarang, jumlahTerjual, totalPendapatan):
        self.__nama = nama
        self.__nomorID = nomorID
        self.__namaUnit = namaUnit
        self.__hargaBarang = hargaBarang
        self.__jumlahTerjual = jumlahTerjual
        self.__totalPendapatan = totalPendapatan
        
    def getNama(self):
        return self.__nama
    
    def getNomorID(self):
        return self.__nama
    
    def getNamaUnit(self):
        return self.__namaUnit
    
    def getHargaBarang(self):
        return self.__hargaBarang
    
    def getJumlahTerjual(self):
        return self.__jumlahTerjual
    
    def getTotalPendapatan(self):
        return self.__totalPendapatan
    
    def getJumlahPenjualan(self):
        return self.__hargaBarang * self.__jumlahTerjual
    
    def getDistributor1(self):
        print("Nama :", self.getNama())
        print("ID : ", self.getNomorID())
        print("Unit : ", self.getNamaUnit())
        self.t = self.getTotalPendapatan() + self.getJumlahPenjualan()
        print("Total Pendapatan :", self.t)
        
    
class wholesaler(distributor):
    def __init__(self, nama, nomorID, namaUnit, hargaBarang, jumlahTerjual, totalPendapatan):
        distributor.__init__(self, nama, nomorID, namaUnit, hargaBarang, jumlahTerjual, totalPendapatan)
        
    def __setBonus(self):
        if self.getJumlahTerjual() >= 1000:
            jt = int(self.getJumlahTerjual / 1000)
            b = jt * 250000
            return b
        else:
            return 0
    
    def getDistributor2(self):  
        print("Nama :", self.getNama())
        print("ID : ", self.getNomorID())
        print("Unit : ", self.getNamaUnit())
        self.t = self.getTotalPendapatan() + self.getJumlahPenjualan() + self.__setBonus()
        print("Total Pendapatan :", self.t)
        
    
class retail(distributor):
    def __init__(self, nama, nomorID, namaUnit, hargaBarang, jumlahTerjual, totalPendapatan):
        distributor.__init__(self, nama, nomorID, namaUnit, hargaBarang, jumlahTerjual, totalPendapatan)
    
    def __setDiskon(self):
        diskon = (self.getJumlahPenjualan() * 5) / 100
        return diskon
        
    def getDistributor3(self):       
        print("Nama :", self.getNama())
        print("ID : ", self.getNomorID())
        print("Unit : ", self.getNamaUnit())
        self.t = self.getTotalPendapatan() + self.getJumlahPenjualan() + self.__setDiskon()
        print("Total Pendapatan :", self.t)
    
d1 = distributor("Ismail M.", "1234", "Distributor", 5000, 50, 200000000)
d1.getDistributor1()

print()

d2 = wholesaler("Ilham T.", "4587", "Wholesaler", 5000, 50, 15000000)
d2.getDistributor2()

print()

d3 = retail("Ilyas M.", "9876", "Retail", 5000, 5000, 20000000)
d3.getDistributor3()
    
```
